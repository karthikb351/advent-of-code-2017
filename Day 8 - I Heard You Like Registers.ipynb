
### This is the old fashioned way

````python
map = {}

def check_op(lhs, rhs, op):
    if lhs not in map:
        map[lhs] = 0
    if op == ">":
        return map[lhs] > rhs
    elif op == "<":
        return map[lhs] < rhs
    elif op == ">=":
        return map[lhs] >= rhs
    elif op == "<=":
        return map[lhs] <= rhs
    elif op == "==":
        return map[lhs] == rhs
    elif op == "!=":
        return map[lhs] != rhs
    else:
        return False
        
max_v = None

for line in input.splitlines():
    parts = line.split(" ")
    l = parts[:]
    op_lhs = parts[0]
    op_rhs = int(parts[2])
    op_op = parts[1]
    c_lhs = parts[4]
    c_op = parts[5]
    c_rhs = int(parts[6])
    
    if check_op(c_lhs, c_rhs, c_op):
        if op_op == "inc":
            v = map[op_lhs] if op_lhs in map.keys() else 0
            map[op_lhs]=v+op_rhs
        else:
            v = map[op_lhs] if op_lhs in map.keys() else 0
            map[op_lhs]=v-op_rhs
        if max_v == None:
            max_v = map[op_lhs]
        max_v = max(map[op_lhs], max_v)
````
